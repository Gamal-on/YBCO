```python
# Déterminer Cmax et Cmin

def Cmax(x, y) :
    idx_max = 0
    Cmax = y[0]
    for i in range(0, len(x)-1) :
        if y[i+1] > Cmax :
            Cmax = y[i+1]
            idx_max += 1 
    return Cmax, x[idx_max]

print(Cmax(temperature_bounded, C_div_T_bounded_temp))
```
